```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

# 1) Prepare the data
# — assume `data` already has Date_N (datetime), Year, Minutes
df = data.dropna(subset=['Year','Minutes','Category']).copy()

# Encode target: Movie=1, TV Show=0
df['is_movie'] = (df['Category']=='Movie').astype(int)

X = df[['Year','Minutes']].values
y = df['is_movie'].values

# 2) Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 3) Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# 4) Fit a logistic regression
clf = LogisticRegression()
clf.fit(X_train_scaled, y_train)

# 5) Evaluate
y_pred = clf.predict(X_test_scaled)
print(classification_report(y_test, y_pred, target_names=['TV Show','Movie']))

# 6) Plot confusion matrix
disp = ConfusionMatrixDisplay.from_estimator(
    clf, X_test_scaled, y_test,
    display_labels=['TV Show','Movie'],
    cmap=plt.cm.Blues,
    normalize=None
)
disp.ax_.set_title("Confusion Matrix")
plt.show()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report,
    RocCurveDisplay,
    ConfusionMatrixDisplay
)

# --- 1) Prepare DataFrame ---
df = data.copy()

# Numeric features
df['Year']    = df['Date_N'].dt.year
df[['Minutes','Unit']] = df['Duration'].str.split(' ', expand=True)
df['Minutes'] = pd.to_numeric(df['Minutes'], errors='coerce')

# Drop rows with missing core features
df = df.dropna(subset=['Year','Minutes','Category','Rating','Country'])

# Target
df['is_movie'] = (df['Category']=='Movie').astype(int)

# Simplify Country into top‑5 + “Other”
country_counts = (
    df['Country']
      .str.split(',', expand=True)
      .stack()
      .str.strip()
      .value_counts()
)
top5 = country_counts.head(5).index.tolist()

def map_country(cell):
    countries = [c.strip() for c in cell.split(',')]
    for c in countries:
        if c in top5:
            return c
    return 'Other'

df['TopCountry'] = df['Country'].apply(map_country)

# --- 2) Train/Test Split ---
X = df[['Year','Minutes','Rating','TopCountry']]
y = df['is_movie']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# --- 3) Build Preprocessing + Model Pipeline ---
numeric_feats = ['Year','Minutes']
numeric_transformer = StandardScaler()

categorical_feats = ['Rating','TopCountry']
categorical_transformer = OneHotEncoder(handle_unknown='ignore', sparse=False)

preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_feats),
    ('cat', categorical_transformer, categorical_feats),
])

pipeline = Pipeline([
    ('prep', preprocessor),
    ('clf', RandomForestClassifier(random_state=42))
])

# --- 4) Hyperparameter Tuning ---
param_grid = {
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [None, 10, 20],
    'clf__min_samples_split': [2, 5],
}

grid = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)
grid.fit(X_train, y_train)

print("Best params:", grid.best_params_)
print("Best CV AUC:", grid.best_score_)

# --- 5) Evaluation on Test Set ---
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:,1]

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['TV Show','Movie']))

# Confusion Matrix
ConfusionMatrixDisplay.from_estimator(
    best_model, X_test, y_test,
    display_labels=['TV Show','Movie']
)
plt.title("Confusion Matrix")

# ROC Curve
RocCurveDisplay.from_predictions(y_test, y_proba, name="RandomForest")
plt.title("ROC Curve")
plt.show()

# --- 6) Feature Importances ---
# Get feature names after one‑hot
ohe = best_model.named_steps['prep'].named_transformers_['cat']
ohe_feats = ohe.get_feature_names_out(categorical_feats)
feat_names = numeric_feats + list(ohe_feats)

importances = best_model.named_steps['clf'].feature_importances_
feat_imp = pd.Series(importances, index=feat_names).sort_values(ascending=False)

plt.figure(figsize=(8,4))
feat_imp.head(10).plot(kind='barh')
plt.title("Top 10 Feature Importances")
plt.xlabel("Importance")
plt.gca().invert_yaxis()
plt.show()
```
