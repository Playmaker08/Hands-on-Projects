```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve

# ---- 1) Data Cleaning & Feature Engineering ----

# Convert Release_Date (object) to datetime and store it as Date_N
data['Date_N'] = pd.to_datetime(data['Release_Date'].str.strip(), errors='coerce')

# Extract Year from Date_N; fill missing years with 0 and convert to int
data['Year'] = data['Date_N'].dt.year.fillna(0).astype(int)

# Minutes is already present as int64 (extracted from Duration)
# If you haven't already computed Minutes and Unit, you'd do:
# data[['Minutes', 'Unit']] = data['Duration'].str.split(' ', expand=True)
# data['Minutes'] = pd.to_numeric(data['Minutes'], errors='coerce').fillna(0)

# Create a new feature: Number of Countries (count comma-separated entries)
# If Country is empty, assign 0
data['Num_Countries'] = data['Country'].fillna("").apply(lambda x: 0 if x == "" else x.count(",") + 1)

# Create target: IsMovie = 1 if Category equals "Movie", else 0
data['IsMovie'] = (data['Category'] == 'Movie').astype(int)

# ---- 2) Prepare Features & Target for ML ----

# Use Year, Minutes, and Num_Countries as features
features = ['Year', 'Minutes', 'Num_Countries']
# Drop rows with missing target values (should be minimal)
df_ml = data.dropna(subset=['Category'])[features + ['IsMovie']]

X = df_ml[features]
y = df_ml['IsMovie']

# ---- 3) Train/Test Split ----

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# ---- 4) Fit Logistic Regression Model ----

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# ---- 5) Predictions & Evaluation ----

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

acc = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_proba)

print(f"Test Accuracy: {acc:.3f}")
print(f"Test ROC-AUC:  {auc:.3f}")

# ---- 6) Plot ROC Curve ----

fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve: Movie vs. TV Show")
plt.legend(loc='lower right')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score

# ——— 1) Prepare raw DataFrame ———
# Assume `data` already has Release_Date parsed to Date_N, Minutes, etc.

# Fill missing Date_N and Minutes if any
data['Date_N'] = pd.to_datetime(data['Release_Date'].str.strip(), errors='coerce')
data['Minutes'] = pd.to_numeric(data['Minutes'], errors='coerce').fillna(0)

# Create features we’ll need inside the pipeline
data['CategoryFlag'] = (data['Category']=='Movie').astype(int)

# ——— 2) Train/Test Split ———
X_raw = data[['Date_N', 'Minutes', 'Country']].copy()
y = data['CategoryFlag']

X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X_raw, y, test_size=0.3, random_state=42, stratify=y
)

# ——— 3) Build a feature‑engineering transformer ———
def featurize(df):
    """Takes a DataFrame with Date_N, Minutes, Country; returns numeric features."""
    df = df.copy()
    # Year
    df['Year'] = df['Date_N'].dt.year.fillna(0).astype(int)
    # Num_Countries
    df['Num_Countries'] = df['Country'].fillna("").apply(lambda x: 0 if x=="" else x.count(",")+1)
    return df[['Year','Minutes','Num_Countries']]

featurizer = FunctionTransformer(featurize, validate=False)

# ——— 4) Build pipeline ———
pipeline = Pipeline([
    ('feats', featurizer),
    ('scale', StandardScaler()),
    ('rf', RandomForestClassifier(random_state=42))
])

# ——— 5) Hyperparameter tuning with GridSearchCV ———
param_grid = {
    'rf__n_estimators': [50, 100, 200],
    'rf__max_depth': [None, 5, 10],
}

grid = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train_raw, y_train)

print("Best params:", grid.best_params_)
print("Best CV accuracy:", grid.best_score_)

# ——— 6) Evaluate on test set ———
y_pred = grid.predict(X_test_raw)
test_acc = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {test_acc:.3f}")

# ——— 7) Feature importances ———
importances = grid.best_estimator_.named_steps['rf'].feature_importances_
feat_names = ['Year','Minutes','Num_Countries']
feat_imp = pd.Series(importances, index=feat_names).sort_values(ascending=False)
print("\nFeature Importances:")
print(feat_imp)

# Plot importances
plt.figure(figsize=(5,3))
feat_imp.plot(kind='bar')
plt.title("Feature Importances")
plt.ylabel("Importance")
plt.tight_layout()
plt.show()

# ——— 8) Confusion matrix ———
cm = confusion_matrix(y_test, y_pred, labels=[1,0])
disp = ConfusionMatrixDisplay(cm, display_labels=['Movie','TV Show'])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix on Test Set")
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import matplotlib.pyplot as plt

# — 3.1) Numeric features from our featurize function
num_feats = featurize(X_raw[['Date_N','Minutes','Country']])
num_feats['Director_Count'] = X_raw['Director_Count']
num_feats['Cast_Count']     = X_raw['Cast_Count']

# — 3.2) TF‑IDF on Description, then reduce dimensionality
tfidf = TfidfVectorizer(max_features=1000, stop_words='english')
desc_tfidf = tfidf.fit_transform(X_raw['Description'])

# Truncate to 50 dimensions for clustering
svd = TruncatedSVD(n_components=50, random_state=42)
desc_reduced = svd.fit_transform(desc_tfidf)  # shape (n_samples, 50)

# — 3.3) Stack numeric + text embeddings into a dense array
num_array  = num_feats.values            # shape (n_samples, 5)
text_array = desc_reduced                # shape (n_samples, 50)
X_cluster  = np.hstack([num_array, text_array])  # shape (n_samples, 55)

# — 3.4) Run K‑Means for k=2…6 and plot inertia (“elbow curve”)
inertias = []
K_range = range(2, 7)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_cluster)
    inertias.append(km.inertia_)

plt.figure(figsize=(6,4))
plt.plot(K_range, inertias, marker='o')
plt.title("K‑Means Elbow Curve")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, confusion_matrix

# — 4.1) Prepare raw features & target
# Assume `data` already has Date_N, Minutes, Num_Countries,
# Director_Count, Cast_Count, and CategoryFlag

X = data[[
    'Year',
    'Minutes',
    'Num_Countries',
    'Director_Count',
    'Cast_Count',
    'Description'
]].copy()

y = data['CategoryFlag']  # 1 for Movie, 0 for TV Show

# — 4.2) Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# — 4.3) ColumnTransformer for numeric + text
preprocessor = ColumnTransformer([
    (
        "num", 
        StandardScaler(), 
        ['Year','Minutes','Num_Countries','Director_Count','Cast_Count']
    ),
    (
        "txt",
        TfidfVectorizer(max_features=500, stop_words='english'),
        'Description'
    )
])

# — 4.4) Full pipeline: preprocessing + RandomForest
full_pipe = Pipeline([
    ('pre', preprocessor),
    ('rf',  RandomForestClassifier(n_estimators=100, random_state=42))
])

# — 4.5) Fit and evaluate
full_pipe.fit(X_train, y_train)
y_pred = full_pipe.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {acc:.3f}")

# — 4.6) Confusion Matrix
cm = confusion_matrix(y_test, y_pred, labels=[1,0])
disp = ConfusionMatrixDisplay(cm, display_labels=['Movie','TV Show'])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 1) Numeric features
features = ['Year','Minutes','Num_Countries','Director_Count','Cast_Count']

# 2) PCA
X = data[features].fillna(0)
X_scaled = StandardScaler().fit_transform(X)
pca = PCA(n_components=2, random_state=42)
pcs = pca.fit_transform(X_scaled)
data['PC1'], data['PC2'] = pcs[:,0], pcs[:,1]

# 3) Correlation
corr = data[features].corr()

# 4) Yearly aggregates
yearly_counts = data.groupby(['Year','Category']).size().reset_index(name='Count')
yearly_avg    = data.groupby('Year')['Minutes'].mean().reset_index(name='Avg_Minutes')

# 5) Plot with extra spacing
fig = plt.figure(figsize=(20, 16))
gs = fig.add_gridspec(
    3, 2,
    left=0.05, right=0.95,
    top=0.95, bottom=0.05,
    hspace=0.6, wspace=0.4
)

# A) PCA scatter (top row)
ax1 = fig.add_subplot(gs[0, :])
sns.scatterplot(
    data=data, x='PC1', y='PC2',
    hue='Category', alpha=0.5, s=60, ax=ax1
)
ax1.set_title('PCA of Numeric Features', fontsize=16, pad=20)
ax1.legend(title='Category', loc='upper right')

# B) Correlation heatmap
ax2 = fig.add_subplot(gs[1, 0])
sns.heatmap(corr, annot=True, fmt=".2f", linewidths=1, square=True, ax=ax2)
ax2.set_title('Feature Correlation Matrix', fontsize=14, pad=12)

# C) Yearly release counts by Category
ax3 = fig.add_subplot(gs[1, 1])
sns.lineplot(
    data=yearly_counts,
    x='Year', y='Count', hue='Category',
    marker='o', linewidth=2.5, markersize=8, ax=ax3
)
ax3.set_title('Yearly Release Counts by Category', fontsize=14, pad=12)
ax3.set_xlim(2005, 2025)
ax3.set_xlabel('Year', labelpad=10)
ax3.set_ylabel('Count', labelpad=10)
ax3.legend(title='Category', loc='upper left')

# D) Yearly average duration
ax4 = fig.add_subplot(gs[2, 0])
sns.barplot(
    data=yearly_avg,
    x='Year', y='Avg_Minutes',
    edgecolor='black', ax=ax4
)
ax4.set_title('Average Title Duration per Year', fontsize=14, pad=12)
ax4.set_ylabel('Avg Duration (min)', labelpad=10)
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=45, ha='right')

# E) Distribution of Cast_Count
ax5 = fig.add_subplot(gs[2, 1])
sns.histplot(
    data=data,
    x='Cast_Count',
    bins=30,
    kde=True,
    edgecolor='black',
    ax=ax5
)
ax5.set_title('Distribution of Cast Sizes', fontsize=14, pad=12)
ax5.set_xlabel('Number of Cast Members', labelpad=10)

plt.show()
```
